```python
import pandas as pd
import numpy as np
```

```python
gdp_raw=pd.read_excel('gdrive/My Drive/Colab Notebooks/wb_gdp_data.xlsx')
emissions_raw=pd.read_excel('gdrive/My Drive/Colab Notebooks/co2emissions.xlsx')
```

```python
# GDP By Country for 2013 to 2023.

gdp_years=pd.DataFrame(gdp_raw.iloc[0:265, 57:69])
countries=pd.DataFrame(gdp_raw.iloc[0:265,0])

gdp=pd.concat([countries, gdp_years], axis=1)
gdp=gdp.rename(columns={"Country Name": "Refrence area"})
gdp
```

```python
# Emissions by Country in 2013.

emissions_2013=pd.DataFrame(emissions_raw[emissions_raw["TIME_PERIOD"] == "2013"])
country_emissions_2013=pd.DataFrame(emissions_2013.loc[:,"Reference area"])
value_counts_2013=pd.DataFrame(emissions_2013.loc[:,"OBS_VALUE"])
emissions_2013=pd.concat([country_emissions_2013, value_counts_2013], axis=1)
emissions_2013=emissions_2013.rename(columns={"OBS_VALUE": "2013"})
emissions_2013

```

```python
# Emissions by Country in 2020.

emissions_2020=pd.DataFrame(emissions_raw[emissions_raw["TIME_PERIOD"] == "2020"])
country_emissions_2020=pd.DataFrame(emissions_2020.loc[:,"Reference area"])
value_counts_2020=pd.DataFrame(emissions_2020.loc[:,"OBS_VALUE"])
emissions_2020=pd.concat([country_emissions_2020, value_counts_2020], axis=1)
emissions_2020=emissions_2020.rename(columns={"OBS_VALUE": "2020"})
emissions_2020
```

```python
# Emissions by Country in 2023.

emissions_2023=pd.DataFrame(emissions_raw[emissions_raw["TIME_PERIOD"] == "2023"])
country_emissions_2023=pd.DataFrame(emissions_2023.loc[:,"Reference area"])
value_counts_2023=pd.DataFrame(emissions_2023.loc[:,"OBS_VALUE"])
emissions_2023=pd.concat([country_emissions_2023, value_counts_2023], axis=1)
emissions_2023=emissions_2023.rename(columns={"OBS_VALUE": "2023"})
emissions_2023
```

```python
emissions_2013_sorted=pd.DataFrame(emissions_2013.sort_values(by= '2013', ascending=False))
emissions_2013.head(10)
```

```python
# Exporting final datasets to .csv

emissions_2013.to_csv("emissions_2013.csv", index=False)
emissions_2020.to_csv("emissions_2020.csv", index=False)
emissions_2023.to_csv("emissions_2023.csv", index=False)
gdp.to_csv("gdp.csv", index=False)
```
